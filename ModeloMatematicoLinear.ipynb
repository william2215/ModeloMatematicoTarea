```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
class Modelo:
  def __init__(self, x, y):
    data = {"x": x, "y": y}
    self.__data =  pd.DataFrame(data)
    self.__x = self.__data.iloc[:,:-1]
    self.__y = self.__data.iloc[:,-1]
    self.__model = None

  @property
  def model_regression(self):
    return self.__model

  def entrenar(self):
    self.__model = LinearRegression()
    self.__model.fit(self.__x,self.__y)
  
  def graficar(self):
    fig, ax = plt.subplots(figsize=(8,5), dpi=100)
    ax.set_title('Modelo de Regresion Lineal')
    ax.set_xlabel('Megabytes')
    ax.set_ylabel('Memory')
    ax = plt.scatter(self.__x, self.__y, s=100)
    ax = plt.plot(self.__x, self.__model.predict(self.__x), color='red')
    plt.show()
  

```

```python

#Datos para probar el modelo matematico

x_test = np.array([3038,8991,7632,3421,4566,9491,4539,8146,8029,3400,5550,8774,5589,3256,6450,5795,4943,4546,3475,6334,3236,6169])
            
y_test = np.array([7541, 22499, 19011, 8599, 11391, 24531, 11349, 20366, 20074, 8501, 13876, 21936, 13974, 8141, 16126, 14489, 12359, 11366, 8689, 15836, 8091, 15424])
```

```python
modelo = Modelo(x_test, y_test)
modelo.entrenar()
```
